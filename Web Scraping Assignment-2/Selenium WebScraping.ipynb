```python
#we get the list of jobs in the left and more details about that job in the right so we need to look at left panal for jobs
#extract jobtitle, company name, yrs of experience and location of the job
leftPanel=driver.find_element(By.CLASS_NAME, "jobCard_jobCard__jjUmu")

#create an empty array
jobTitles=[]
companies=[]
experiences=[]
locations=[]

#search for titles of the job
titles=leftPanel.find_elements(By.XPATH, "//h2")
#loop through each jobs and get the text value and insert into the list of array
for t in titles:
    title=t.text
    jobTitles.append(title)
    
companyNames=leftPanel.find_elements(By.XPATH, "//div[@class='jobCard_jobCard_cName__mYnow']")
for c in companyNames:
    company=c.text
    companies.append(company) 
    
experience=driver.find_elements(By.CLASS_NAME, "jobCard_jobIcon__3FB1t")
for e in experience:
    exp=e.text
    experiences.append(exp) 
    
location=driver.find_elements(By.CLASS_NAME, "jobCard_locationIcon__zrWt2")
for l in location:
    place=l.text
    locations.append(place)   

#because all the columns are of different size. DataFrame expects all the array to be of same size.
data={'JobTitle':jobTitles, 'CompanyName': companies, 'Experience':experiences, 'Location':locations}
df = pd.DataFrame.from_dict(data, orient='index')
df = df.transpose()
df.head(10)
```

```python
#we have 40 items in each page
start=0
end=3

brand_names=[]
product_descriptions=[]
prices=[]
offers=[]

for page in range(start,end):
    brands=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for b in brands:
        brand=b.text
        brand_names.append(brand)
        
    products=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for p in products:
        product=p.text
        product_descriptions.append(product)
        
    values=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for v in values:
        value=v.text
        prices.append(value)

    off=driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    for o in off:
        offer=o.text[:-3]
        offers.append(offer) 
        
    next_page=driver.find_element(By.XPATH,"//a[@class='ge-49M']")
    next_page.click()
    time.sleep(3) #time.sleep(3) takes few secs to move        
      
#print(len(brand_names), len(product_descriptions)) to see what is the length of each columns

data={'Brand':brand_names, 'Product':product_descriptions, 'Price':prices, 'Offer':offers}
df = pd.DataFrame.from_dict(data, orient='index')
df = df.transpose()
df.head(100)
```

```python
#here we are extracting the data of Rating, summary and reviews of the Apple Phone
#there are 10 results in each page 
start=0
end=10

ratings=[]
rSummary=[]
reviews=[]

for page in range(start,end): #range(0,10)
    rating=driver.find_elements(By.XPATH,'//div[@class="_3LWZlK"]')
    for r in rating:
        star=r.text
        ratings.append(star)

    summary=driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    for s in summary:
        summ=s.text
        rSummary.append(summ)
        
    review=driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
    for rw in review:
        r=rw.text
        reviews.append(r)
        
    
data={'Ratings':ratings,'Summary':rSummary,'Review':reviews}
df = pd.DataFrame.from_dict(data, orient='index')
df = df.transpose()
df.head(100)
```
